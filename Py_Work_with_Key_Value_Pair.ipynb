```python
import numpy as np
import ast as ast

#--declaring list of years
List_Year = list(range(2000,2010))

print("List_Year --> " + str(List_Year))
print(" ")

print("Type of List_Year --> " + str(type(List_Year)))
print(" ")

List_Year_Len = len(List_Year)
print("Length of List_Year --> " + str(List_Year_Len))
print(" ")

dict_content = "{"

for i in range(0,List_Year_Len):
    if (i == List_Year_Len - 1):
        dict_content = dict_content + "'" + str(List_Year[i]) + "'" + ":" + str(i) + "}"
    else:
        dict_content = dict_content + "'" + str(List_Year[i]) + "'" + ":" + str(i) + ","

print("dict_content --> " + str(dict_content))
print(" ")
        
#--declaring dictionary of years
Dict_Year_Col = ast.literal_eval(dict_content) 

print("Dict_Year_Col --> " + str(Dict_Year_Col))
print(" ")

print("Type of Dict_Year_Col --> " + str(type(Dict_Year_Col)))
print(" ")

print("Dict_Year_Col['2000'] --> " + str(Dict_Year_Col['2000']))
print(" ")
```

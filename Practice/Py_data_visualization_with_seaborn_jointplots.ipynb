```python
##--> importing packages
import pandas as pd
import numpy as np

##--> importing data from csv file to the data frame
movie_ratings = pd.read_csv("C://Users//dbandyopadhyay//Documents//Sample_Data//Movie-Ratings.csv")

##--> checking top 5 rows from data frame
print("top five rows in movie ratings : ")
print(movie_ratings.head())

##--> checking bottom 5 rows from data frame
print("bottom five rows in movie ratings : ")
print(movie_ratings.tail())
```

```python
##--> checking the data count
print("total movie ratings : ") 
print(str(movie_ratings.count()))
```

```python
##--> checking the column names
print(str(movie_ratings.columns))
```

```python
##--> modifying the column names
movie_ratings.columns = ["Film","Genre","CriticRating","AudienceRating","BudgetInMillions","Year"]
movie_ratings.columns
```

```python
##--> analyzing the movie ratings data frame
movie_ratings.describe()

##--> data from year columns is not getting treated as year. No we need categorize the data
```

```python
##--> modifying the category of data for columns --> File, Genre, Year
movie_ratings.Film = movie_ratings.Film.astype("category")
movie_ratings.Genre = movie_ratings.Genre.astype("category")
movie_ratings.Year = movie_ratings.Year.astype("category")
```

```python
##--> checking data from FIlm column of movie_ratings data frame
print(movie_ratings.Film)
print(" ")
print(" ")
print(movie_ratings.Film.cat.categories)
```

```python
##--> checking data from Genre column of movie_ratings data frame
print(movie_ratings.Genre)
print(" ")
print(" ")
print(movie_ratings.Genre.cat.categories)
```

```python
##--> checking data from Year column of movie_ratings data frame
print(movie_ratings.Year)
print(" ")
print(" ")
print(movie_ratings.Year.cat.categories)
```

```python
##--> checking information about movie_ratings
movie_ratings.info()

##--> You can see the datatype for columns --> Film, Genre, Year is different and that is category  
```

```python
##--> analyzing the movie_ratings data frame again
movie_ratings.describe()

##--> You can see that the data from Year column is not considered below for analysis as now it has a datatype of category
```

```python
##--> How to check the unique values in Genre and Year columns in movie_ratings data frame
print("Unique values in genre column : ")
print(movie_ratings.Genre.cat.categories)
print(" ")
print(" ")
print("Unique values in year column : ")
print(movie_ratings.Year.cat.categories)
```

```python
##--> Creating joint plot
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
import seaborn as sns
import warnings
%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
##--> adding default gridline for seaborn plot
sns.set()
```

```python
##--> plotting the jointplot with default attributes
jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating")

##--> the grid lines are visible as well as scatter plot and bar charts for audiencertating and criticrating
```

```python
##--> setting grid line style
sns.set_style("whitegrid")

##--> plotting the jointplot with regression analysis
jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", kind="reg")
```

```python
##--> plotting the jointplot with regression analysis with different color

jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", kind="reg", color="g")

jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", kind="reg", color="r")

jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", kind="reg", color="k")
```

```python
##--> plotting the jointplot with hexagons
jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", kind="hex", color="r")
```

```python
##--> plotting the jointplot with kde
jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", kind="kde", color="r")
```

```python
##--> plotting the jointplot with resid
jp1 = sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", color="r", kind="resid" )
```

```python
##--> plotting the jointplot with kde and doing cosmetic changes and adding custome labels
jp1 = (sns.jointplot(data=movie_ratings,x="CriticRating",y="AudienceRating", color="r", kind="kde", edgecolor="w", s=40)\
       .set_axis_labels("audience ratings","critic ratings"))  
```

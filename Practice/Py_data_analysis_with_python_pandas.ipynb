```python
##--> importing required packages for data analysis
from IPython.core.interactiveshell import InteractiveShell
import pandas as pd
import numpy as np
InteractiveShell.ast_node_interactivity = "all"
# InteractiveShell.ast_node_interactivity = "last_expr"
```

```python
##--> importing data from csv file to pandas data frame
mr1 = pd.read_csv("C://Users//dbandyopadhyay//Documents//Sample_Data//Movie-Ratings.csv")

print("=============================================================================================================")
print(" ")
print("Checking first five rows in the data frame : ")
mr1.head()
print(" ")

print("=============================================================================================================")
print(" ")
print("Checking last five rows in the data frame : ")
mr1.tail()
print(" ")

print("=============================================================================================================")
print(" ")
print("Checking column names in the data frame : ")
mr1.columns
print(" ")

print("=============================================================================================================")
print(" ")
print("modifying column names of the data frame : ")
mr1.columns = ["film","genre","critic_rating","audience_rating","budget_in_millions","release_year"]
mr1.columns
print(" ")

print("=============================================================================================================")
print(" ")
print("Checking datatypes of columns in the data frame : ")
mr1.dtypes
print(" ")

print("=============================================================================================================")
print(" ")
print("Changing data to category data for columns film, critic_rating, audience_rating  in the data frame : ")
mr1["film"] = mr1["film"].astype("category")
mr1["genre"] = mr1["genre"].astype("category")
mr1["release_year"] = mr1["release_year"].astype("category")
mr1.dtypes
print(" ")

print("=============================================================================================================")
print(" ")
print("describing the movie ratings data frame : ")
mr1.describe()
print(" ")

```

```python
##--> grouping the data to check count, sum, min, max etc
print("=============================================================================================================")
print(" ")
print("Total no of movies released per year : ")
print(mr1.groupby(["release_year"])[["budget_in_millions"]].count())
print(" ")

print("=============================================================================================================")
print(" ")
print("Total no of movies released per genre: ")
print(mr1.groupby(["genre"])[["budget_in_millions"]].count())
print(" ")
```

```python
print("=============================================================================================================")
print(" ")
print("Total no of movies released per year per genre: ")
print(mr1.groupby(["release_year","genre"])[["budget_in_millions"]].count())
print(" ")
```

```python
print("=============================================================================================================")
print(" ")
print("Sum of movie budgets released per year per genre: ")
print(mr1.groupby(["release_year","genre"])[["budget_in_millions"]].sum())
print(" ")
```

```python
print("=============================================================================================================")
print(" ")
print("Checking movie with max budget released per year per genre: ")
print(mr1.groupby(["release_year","genre"])[["budget_in_millions"]].max())
print(" ")
```

```python
print("=============================================================================================================")
print(" ")
print("Checking movie with max budget released per year per genre: ")
print(mr1.groupby(["release_year","genre"])[["budget_in_millions"]].min())
print(" ")
```

```python
print("=============================================================================================================")
print(" ")
print("Checking movie with max budget released per year per genre: ")
print(mr1.groupby(["release_year","genre"])[["budget_in_millions"]].prod())
print(" ")
```

```python
print("=============================================================================================================")
print(" ")
df_bim = mr1.groupby(["release_year","genre"],as_index=False).agg({"budget_in_millions" : "sum"})
df_bim
print(" ")
```

```python
print("=============================================================================================================")
print(" ")
df_cnt = mr1.groupby(["release_year","genre"],as_index=False).agg({"film" : "count"})
df_cnt
print(" ")
```

```python
df_bim_cnt = pd.merge(df_bim, df_cnt, how = 'inner', left_on = ["release_year","genre"], right_on = ["release_year","genre"])
df_bim_cnt
```

```python
df_bim_cnt_1 = mr1.groupby(["release_year","genre"],as_index=False).agg({"budget_in_millions" : "sum","film" : "count"})
df_bim_cnt_1
```

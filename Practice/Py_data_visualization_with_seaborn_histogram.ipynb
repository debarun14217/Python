```python
print("--------------------------------------------------------------------------------------------------")
print("Plotting histogram with bins from a list of genre: " )
genre_list = list()
genre_label_list = list()

genre_list.append(movie_ratings[movie_ratings.Genre == "Comedy"].BudgetInMillions)
genre_label_list.append("Comedy")

genre_list.append(movie_ratings[movie_ratings.Genre == "Action"].BudgetInMillions)
genre_label_list.append("Action")

genre_list.append(movie_ratings[movie_ratings.Genre == "Adventure"].BudgetInMillions)
genre_label_list.append("Adventure")

genre_list.append(movie_ratings[movie_ratings.Genre == "Drama"].BudgetInMillions)
genre_label_list.append("Drama")

genre_list.append(movie_ratings[movie_ratings.Genre == "Horror"].BudgetInMillions)
genre_label_list.append("Horror")

hist = plt.hist(genre_list, bins = 15, label = genre_label_list, rwidth = 1)
hist = plt.legend()
print(" ")
```

```python
print("--------------------------------------------------------------------------------------------------")
print("Plotting stacked histogram with bins from a list of genre: " )
genre_list = list()
genre_label_list = list()

genre_list.append(movie_ratings[movie_ratings.Genre == "Comedy"].BudgetInMillions)
genre_label_list.append("Comedy")

genre_list.append(movie_ratings[movie_ratings.Genre == "Action"].BudgetInMillions)
genre_label_list.append("Action")

genre_list.append(movie_ratings[movie_ratings.Genre == "Adventure"].BudgetInMillions)
genre_label_list.append("Adventure")

genre_list.append(movie_ratings[movie_ratings.Genre == "Drama"].BudgetInMillions)
genre_label_list.append("Drama")

genre_list.append(movie_ratings[movie_ratings.Genre == "Horror"].BudgetInMillions)
genre_label_list.append("Horror")

hist = plt.hist(genre_list, bins = 15, label = genre_label_list, rwidth = 1, stacked = True)
hist = plt.legend()
print(" ")
```

```python
print("--------------------------------------------------------------------------------------------------")
print("Plotting stacked histogram by using automation : " )

genre_list = list()
genre_label_list = list()

for genre in movie_ratings.Genre.cat.categories:
    #print(genre)
    #print(movie_ratings[movie_ratings.Genre == genre].BudgetInMillions)
    genre_list.append(movie_ratings[movie_ratings.Genre == genre].BudgetInMillions)
    genre_label_list.append(genre)

hist = plt.hist(genre_list, bins = 15, label = genre_label_list, rwidth = 1, stacked = True)
hist = plt.legend()
print(" ")
```

```python
print("--------------------------------------------------------------------------------------------------")
print("Plotting stacked histogram with bins from a list of years and automating the creation of list : " )

year_list = list()
year_label_list = list()

for year in movie_ratings.Year.cat.categories:
    #print(genre)
    #print(movie_ratings[movie_ratings.Genre == genre].BudgetInMillions)
    year_list.append(movie_ratings[movie_ratings.Year == year].BudgetInMillions)
    year_label_list.append(year)

hist = plt.hist(year_list, bins = 15, label = year_label_list, rwidth = 1, stacked = True)
hist = plt.legend()
print(" ")
```

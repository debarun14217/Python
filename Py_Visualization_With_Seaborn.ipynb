```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline
plt.rcParams["figure.figsize"] = 16,8
```

```python
##--> importing data from CSV file
df1 = pd.read_csv("C:\\Users\\DEBARUN\\Documents\\Python Scripts\\Data_sets\\DemographicData.csv")
df1.head()
```

```python
##--> Checking max value of internet user
df1["Internet users"].max()
```

```python
##--> distplot creation based n internet user data
vis1 = sbn.distplot(df1["Internet users"],bins = 30,color = "r")
sbn.set_style("whitegrid", {'axes.grid' : True})
```

```python
##--> distplot creation based n birth rate data
vis2 = sbn.distplot(df1["Birth rate"], bins = 25, color = "green")
```

```python
##--> boxplot creation based on income group and birth rate data
vis3 = sbn.boxplot(data = df1, x="Income Group", y="Birth rate")
```

```python
##--> trying plot linear regression
vis4 = sbn.lmplot(data = df1, x = 'Internet users', y = 'Birth rate', fit_reg = True, \
                  hue ="Income Group",size = 8, aspect = 1)
```

```python
##--> removing linear regression analysis
vis4 = sbn.lmplot(data = df1, x = 'Internet users', y = 'Birth rate', fit_reg = False, \
                  hue ="Income Group",size = 8, aspect = 1,                            \
                  scatter_kws = {"s" : 100, "marker" : '8'})
```

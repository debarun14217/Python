```python
##--> we can write the code like the following to generate the subplots
f, ax = plt.subplots(2, 2, figsize = (12,10))

kdeplot1 = sns.kdeplot(mv_set.BudgetMillions, mv_set.AudienceRatings, ax=ax[0,0])
kdeplot2 = sns.kdeplot(mv_set.BudgetMillions, mv_set.CriticRatings, ax=ax[1,1])

##--> you can also set the same range for x axis and y axis for all the subplots
kdeplot1.set(xlim=(-20,300))
kdeplot1.set(ylim=(-20,140))
kdeplot2.set(xlim=(-20,300))
kdeplot2.set(ylim=(-20,140))
```

```python
##--> or we can use the same functionalty by using the following code
f, ax = plt.subplots(2, 2, figsize = (12,10), sharex = True, sharey = True)

kdeplot1 = sns.kdeplot(mv_set.BudgetMillions, mv_set.AudienceRatings, ax=ax[0,0])
kdeplot2 = sns.kdeplot(mv_set.BudgetMillions, mv_set.CriticRatings, ax=ax[1,1])
```
